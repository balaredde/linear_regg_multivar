```python
import pandas as pd
import numpy as np
from sklearn import linear_model

```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
a=3
```

```python
df.bedrooms = df.bedrooms.fillna(a)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']], df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```
